```python
#Simple test to poke some registers

from pynq import Overlay

filename= "DPU.bit"

overlay = Overlay("DPU.bit")

print(f"Overlay {filename} loaded...\n")

for ip in overlay.ip_dict.keys():
    print(f"- {ip} \t instantiated")

dpu_keys = [key for key in overlay.ip_dict.keys() if 'dpu' in key.lower()]

if dpu_keys:
    dpu_name = dpu_keys[0]
    
    # Gemini says that this grabs the AXI memory-mapped IP block
    dpu_ip = getattr(overlay, dpu_name)
    
    # 0x00 version reg. 
    version = dpu_ip.read(0x00)
    print(f"\n[{dpu_name}] Version: {hex(version)}")
    
    # 0x04 reset register
    status = dpu_ip.read(0x04)
    print(f"[{dpu_name}] Status: {hex(status)}")
    if hex(status) == hex(15):
        print("\n\n------------------------- ITTY BITTY ERROR -------------------------")
        print("\n [ERROR] Currently in active high reset state. Attempting to correct\n")
        print(f" Writing register 0x04 (reset) to 0x0 (should allow core to run)")
        dpu_ip.write(0x04, 0x0)
        status = dpu_ip.read(0x04) # recheck the value
        if hex(status) != hex(15):
            print(f" [{dpu_name}] New status: {hex(status)}")
            print("\n\n   ---------------- ITTY BITTY ERROR RESOLVED -------------------")
            print("\n [SUCCESS]DPU succesfully written to.\n")
            print("\nDPU running. Yahoo! All tests passed!")

    else:
        print("\n[SUCCESS] DPU reset values correctly set!. Yahoo! All (one) tests passed!")

        
else:
    print("\nDPU not found")
```
